```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV with market data
df = pd.read_csv("orders.csv")

# Plot all instruments in subplots
fig, axes = plt.subplots(5, 2, figsize=(12, 10), sharex=True)
for i, ax in enumerate(axes.flatten()):
    sub = df[df["instrument_id"] == i]
    ax.plot(sub["send_time_ns"], sub["price"])
    ax.set_title(f"Instrument {i}")
plt.suptitle("Price Over Time for Each Instrument (Orders Only)", fontsize=16)
plt.show()
```
